# 04 - Processing Data

#### Dealing with outliers.

```python
# For the moment, I decided not to remove any of the outliers to not lose any observations
# Instead I've removed some columns which did not have any or almost no correlation with the target
```

```python
for c in numerical.columns:
    sns.boxplot(numerical[c])
    plt.show()
```

```python
sns.boxplot(numerical.monthly_premium_auto, whis=4)
```

```python
iqr = np.quantile(numerical.monthly_premium_auto, 0.75) - np.quantile(numerical.monthly_premium_auto, 0.25)
iqr
```

```python
upper_limit = np.quantile(numerical.monthly_premium_auto, 0.75) + 4*iqr
lower_limit = np.quantile(numerical.monthly_premium_auto, 0.25) - 4*iqr
upper_limit, lower_limit
```

```python
# Number of outliers with whis = 4 in the monthly_premium_auto column
len(numerical[numerical.monthly_premium_auto < lower_limit] & numerical[numerical.monthly_premium_auto > upper_limit])
```

```python
# Checking for outliers in the customer lifetime value column
sns.boxplot(numerical.customer_lifetime_value, whis=10)
```

```python
iqr2 = np.percentile(numerical.customer_lifetime_value,75) - np.percentile(numerical.customer_lifetime_value,25)
iqr2
```

```python
up_lim = np.percentile(numerical.customer_lifetime_value, 75) + 10*iqr2
low_lim = np.percentile(numerical.customer_lifetime_value, 25) - 10*iqr2
up_lim, low_lim
```

```python
# Number of outliers with whis = 10 in the customer lifetime value column
len(numerical[numerical.customer_lifetime_value < low_lim] & numerical[numerical.customer_lifetime_value > up_lim])
```

```python
# Drop some numerical columns with +/- 0 - 5% correlation with the target
numerical.columns
```

```python
drop = ['months_since_last_claim', 'months_since_policy_inception',
       'number_of_open_complaints', 'number_of_policies']
num = numerical.drop(drop, axis=1)
num = round(num,2)
num.head()
```

#### Normalization.

```python
# X,y Split before normalizing
y = num.total_claim_amount
X = num.drop('total_claim_amount', axis=1)
```

```python
X.head()
```

```python
transformer = MinMaxScaler().fit(X) #Normalize with MinMaxScaler
X_minmax = transformer.transform(X)
print(X_minmax.shape)
```

```python
X_norm = pd.DataFrame(X_minmax,columns=X.columns)
X_norm.head() 
```

```python
for col in X_norm.columns:
    sns.displot(X_norm[col],kde=True)
    plt.show()
```

#### Encoding Categorical Data.

```python
# Drop columns that are not helpful in determining the target
cat.columns
```

```python
catdrop= ['response', 'education', 'marital_status', 'renew_offer_type', 'sales_channel']
cat1 = cat.drop(catdrop, axis=1)
cat1.info()
```

```python
# One Hot Encoder for Categoricals
cat_dumm = pd.get_dummies(cat1)
cat_dumm.head()
```

```python
# Concatenate numerical and transformed categorical variables
X_num_cat = pd.concat([X_norm, cat_dumm], axis=1)
X_num_cat.head()
```

#### Splitting into train set and test set.

```python
# Train-Test 70/30 Split
X_train, X_test, y_train, y_test = train_test_split(X_num_cat, y, test_size=0.3, random_state=123)
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

# 05 - Modeling
Apply model.

```python
# Apply Linear Regression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# Predictions
predict = model.predict(X_test)
predict
```

# 06 - Model Validation

```python
y.describe()
```

```python
r2 = r2_score(y_test, predict)
RMSE = mean_squared_error(y_test, predict, squared=False)
MSE = mean_squared_error(y_test, predict)
MAE = mean_absolute_error(y_test, predict)
print("r2 = ", r2)
print("RMSE = ", RMSE)
print("MSE = ", MSE)
print("MAE = ", MAE)
```
